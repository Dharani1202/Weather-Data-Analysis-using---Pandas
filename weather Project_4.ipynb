```python
#4.	Find out the number of null values in each columns and remove those null values
# ther is no null value

ds.isnull().sum()
```

```python
#5.	Find all the instances when “snow” is recorded

ds[ds['Weather']=='Snow']
```

```python
#6.	Find out all the instanced when “wind speed is above 24” and “visibility is 25”.

ds[(ds['Wind Speed_km/h']>24)&(ds['Visibility_km']==25)]
```

```python
#7.	Show all records where weather condition is fog

ds[ds['Weather']=="Fog"]
```

```python
#8.	Find all the instance where weather is clear or visibility is Greater than 40

ds[(ds['Weather']=='clear')|(ds['Visibility_km']>=40)]
```

```python
#9.	Find all the instance where weather is clear and relative humidity is greater than 50
```

```python
ds[(ds['Weather']=='Clear')&(ds['Rel Hum_%']>50)]
```
